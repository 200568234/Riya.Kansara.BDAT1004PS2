Now write a class named Kangaroo as a subclass of Marsupial that inherits all the 
attributes of Marsupial and also: 
a. extends the Marsupial __init__ constructor to take, as input, the 
coordinates x and y of the Kangaroo object, 
b. supports method jump that takes number values dx and dy as input and 
moves the kangaroo by dx units along the x-axis and by dy units along the y
axis, and 
c. overloads the __str__ operator so it behaves as shown below. 
 
k = Kangaroo(0,0)

print(k) 

I am a Kangaroo located at coordinates (0,0) 

k.put_in_pouch('doll')

['doll', 'firetruck', 'kitten']

print(k) 

I am a Kangaroo located at coordinates (3,0)


```python
class Marsupial:
    def __init__(self):
        self.pouch = []

    def put_in_pouch(self, item):
        self.pouch.append(item)

    def pouch_contents(self):
        return self.pouch

class Kangaroo(Marsupial):
    def __init__(self, x, y):
        super().__init__()
        self.x = x
        self.y = y

    def jump(self, dx, dy):
        self.x += dx
        self.y += dy

    def __str__(self):
        return f"I am a Kangaroo located at coordinates ({self.x},{self.y})"


m = Marsupial()
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')
print(m.pouch_contents())

k = Kangaroo(0, 0)
print(k)
k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
print(k.pouch_contents())
k.jump(1, 0)
k.jump(1, 0)
k.jump(1, 0)
print(k)
```

### Question 5

Write a recursive method binary() that takes a non-negative 
integer n and prints the binary representation of integer n. 
 
binary(0) 

0 

binary(1)

binary(3)

```python
def binary(n):
    if n == 0:
        print(0, end='')
    elif n == 1:
        print(1, end='')
    else:
        binary(n // 2)
        print(n % 2, end='')

print("binary(0)")
binary(0)
print()  
print("binary(1)")
binary(1)
print()  
print("binary(3)")
binary(3)
print()  
print("binary(9)")
binary(9)
print()  
```

### Question 6
Implement a class named HeadingParser that can be used to parse an HTML document, and retrieve and print all the headings in the document. You should implement your class as a subclass of HTMLParser, defined in Standard Library module html.parser. When fed a string containing HTML code, your class should print the headings, one per line and in the order in which they appear in the 
document. Each heading should be indented as follows: an h1 heading should have indentation 0, and h2 heading should have indentation 1, etc. Test your 
implementation using w3c.html. 
 
infile = open('w3c.html')

content = infile.read() 

infile.close() 

hp = HeadingParser() 

hp.feed(content) 

W3C Mission 

 Principles 

```python
from html.parser import HTMLParser

class HeadingParser(HTMLParser):
    def __init__(self):
        super().__init__()
        self.headings = []
        self.current_level = 0

    def handle_starttag(self, tag, attrs):
        if tag.startswith('h') and len(tag) > 1 and tag[1].isdigit():
            # Set the current level based on the heading tag number
            self.current_level = int(tag[1])

    def handle_endtag(self, tag):
        if tag.startswith('h') and len(tag) > 1 and tag[1].isdigit():
            # Reset the current level after closing the tag
            self.current_level = 0

    def handle_data(self, data):
        if self.current_level > 0 and data.strip():  # Only add non-empty data
            # Append the heading with indentation based on level
            self.headings.append(" " * (self.current_level - 1) + data.strip())

    def print_headings(self):
        for heading in self.headings:
            print(heading)

# Reading the HTML content from 'w3c.html' and parsing
with open('w3c.html', 'r') as f:
    content = f.read()

hp = HeadingParser()
hp.feed(content)
hp.print_headings()
```

Implement recursive function webdir() that takes as input: a URL (as a string) and 
non-negative integers depth and indent. Your function should visit every web 
page reachable from the starting URL web page in depth clicks or less, and print 
each web page's URL. As shown below, indentation, specified by indent,  should 
be used to indicate the depth of a URL. 
 
webdir('http://reed.cs.depaul.edu/lperkovic/csc242/test1.html'
 , 2, 0) 
http://reed.cs.depaul.edu/lperkovic/csc242/test1.html 
    http://reed.cs.depaul.edu/lperkovic/csc242/test2.html 
        http://reed.cs.depaul.edu/lperkovic/csc242/test4.html 
    http://reed.cs.depaul.edu/lperkovic/csc242/test3.html 
        http://reed.cs.depaul.edu/lperkovic/csc242/test4.html

```python
import requests
from bs4 import BeautifulSoup


def get_links(url):
    """Returns a list of links on the given web page"""
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    links = []
    for link in soup.find_all('a'):
        href = link.get('href')
        if href is not None:
            links.append(href)
    return links


def webdir(url, depth, indent=0):
    """Recursively visits web pages reachable from the starting URL
    in depth clicks or less and prints each web page's URL."""
    print(' ' * indent + url)
    if depth == 0:
        return
    links = get_links(url)
    for link in links:
        if link.startswith('http'):
            webdir(link, depth-1, indent+1)
            

webdir('http://reed.cs.depaul.edu/lperkovic/csc242/test1.html', 2, 0)
```
